```python
import pandas as pd 
import numpy as np 

# Librerías de visualización
import seaborn as sns
import matplotlib.pyplot as plt

#Calendario meses
import calendar

# ver todas las columnas
pd.set_option('display.max_columns', None)
# ver todas las filas
pd.set_option('display.max_rows', None)
```

```python
df_customer = pd.read_csv('Customer_data_clean.csv', on_bad_lines='skip', index_col = 0)
```

```python
def eda_basico(df):

    print('🌷Ejemplo de datos del DF:')
    display(df.head(3))
    display(df.tail(3))
    display(df.sample(3))
    print('________________________________________________________________________________________________________')

    print('🌻Número de Filas:')
    display(df.shape[0])
    print('________________________________________________________________________________________________________')

    print('🌱Número de Columnas:')
    display(df.shape[1])
    print('________________________________________________________________________________________________________')

    print('🌼Información de la tabla:')
    display(df.info())
    print('________________________________________________________________________________________________________')

    print('🧬 Tipos de datos por columna:')
    display(df.dtypes.value_counts())
    print('________________________________________________________________________________________________________')

    print('🌑Nombre de las columnas:')
    display(df.columns)
    print('________________________________________________________________________________________________________')

    print('🍄Descripción de los datos numéricos:')
    display(df.describe().T)
    print('________________________________________________________________________________________________________')

    print('🌋Descripción de los datos no-numéricos:')
    if df.select_dtypes(include='object').shape[1] > 0:
        display(df.describe(include='object').T)
    else:
        print('No hay columnas no-numéricas.')
    print('________________________________________________________________________________________________________')

    print('🍂Saber si hay datos únicos:')
    display(df.nunique())
    print('________________________________________________________________________________________________________')

    print('🐖Que datos son nulos por columnas:')
    display(df.isnull().sum())
    print('________________________________________________________________________________________________________')

    print('🐲Filas duplicadas:')
    total_duplicados = df.duplicated().sum()
    if total_duplicados > 0:
        print(f'cantidad de duplicados: {total_duplicados}')
        print('Primeros duplicados')
        display(df[df.duplicated()].head(3))
    else:
        print('No hay duplicados')
    print('________________________________________________________________________________________________________')

    print('🪹 Columnas constantes (solo 1 valor único):')
    constantes = df.columns[df.nunique() <= 1]
    if len(constantes) > 0:
        print(f'{len(constantes)} columnas con 1 valor único:')
        display(constantes)
    else:
        print('No hay columnas constantes')
    print('________________________________________________________________________________________________________')
    
    return
```

```python
eda_basico(df_customer)
```

## VISUALIZACIÓN DE DATOS

```python
#1. ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?

df_mes = df_customer.groupby('Month')['Flights Booked'].sum().sort_index()

df_mes
```

```python
#Visualización de los vuelos reservados por meses:

palette = sns.color_palette("Set2", n_colors=len(df_mes))

plt.figure(figsize=(10, 6))
plt.bar(x=df_mes.index, height=df_mes.values, color= palette)

# Mostrar todos los meses con sus nombres
plt.xticks(ticks=df_mes.index, labels=[calendar.month_name[i] for i in df_mes.index], rotation=45)

plt.title("Cantidad de vuelos reservados por mes")
plt.xlabel('Mes')
plt.ylabel('Número de reservas')
plt.tight_layout()
plt.show()

#Esto nos muestra como los meses fuertes son verano, sobre todo Julio y Diciembre, coincidiendo con las vacaciones de verano y Navidad.
#Como el contexto es Canadá también coincide con los meses de más calor y cuando las familias viajan más ya que las escuela estan cerradas y 
#Diciembre también es cuando empieza a hacer mucho frio y muchos canadienses escapan a Mexico, el Caribe o EEUU.
```

```python
#2. ¿Existe una relación entre la distancia de los vuelos y los puntos acumulados por los cliente?

#Primero agrupamos la distancia y puntos acumulados por cliente ya que salen múltiples veces en el DF:

df_distancia = df_customer.groupby('Loyalty Number')[['Distance', 'Points Accumulated']].sum().reset_index().sort_values(by='Distance', ascending=False)

df_distancia.sample(5)

#Esto ya nos indica que es muy probable que haya una correlación entre las variables
```

```python
#Hago un mapa de correlación (heatmap) para ver más claramente la relación entre ls variables:

correlation = df_distancia[['Distance', 'Points Accumulated']].corr()

plt.figure(figsize = (5, 5))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5,vmin=-1, vmax=1, mask = np.triu(np.ones_like(correlation, dtype=bool)))
plt.title('Matriz de Correlación entre Variables')
plt.show()

# Esto muestra una relación perfecta con 1 punto, lo que indica mayor distancia mas puntos acumulados.
```

```python
sns.scatterplot(data=df_distancia, x='Distance', y='Points Accumulated')
plt.title("Distancia vs Puntos Acumulados")
plt.xlabel("Distancia Total por cliente")
plt.ylabel("Puntos Acumulados")
plt.show()

# Esta linea casi perfecta también confirma la relación entre ambas variables.
```

```python
#3.¿Cuál es la distribución de los clientes por provincia o estado?

#Cuantas provincias hay? 11
df_customer['Province'].unique()
```

```python
#Creo primero una variable para que agrupe a los clientes por su numbero de cliente (Loyalty Number) y pongo en agg que quiero que haga con cada valor.
# Por ejemplo gender o provincia que se quede con el primero que salga pero en flights booked o Points Accumulated necessito que los sume, ya que sino 
# solo cogería el primer valor. 
clientes  = df_customer.groupby('Loyalty Number').agg({'Flights Booked': 'sum', 'Education': 'first', 'Distance': 'sum', 'Points Accumulated': 'sum', 'Province': 'first', 'Salary': 'first', 'Loyalty Card': 'first', 'Marital Status': 'first', 'Gender': 'first'}).reset_index()
```

```python
#Contar cuantos clientes hay por provincia:

df_provincia = clientes['Province'].value_counts().sort_values()

df_provincia
```

```python
#Visualización de la distribución de clientes entre las 11 provincias de Canadá:

plt.figure(figsize=(10, 6))
df_provincia.sort_values().plot(kind='barh', color='purple')
plt.title('Distribución de Clientes por Provincia')
plt.xlabel('Número de Clientes')
plt.ylabel('Provincia')
plt.tight_layout()
plt.show()

#Esto nos muestra que la mayoria de clientes de esta aerolinea viven en Ontario con un total de 5399 personas, 
# la provincia con menor número de clientes es Prince Edward Island. Esto también tiene relación al número de habitantes por cada provincia
# siendo también Ontario la que tiene más habitantes (15.8 millones) y la Isla del Principe Eduardo la que menos (180mil)
```

```python
#4. ¿Cómo se compara el salario promedio entre los diferentes niveles educativos de los clientes?

#Sigo utilizando la variable clientes para que agrupe por cliente unico.

#Calculo promedio del salario:
salary = clientes.groupby('Education')['Salary'].mean().round().sort_values().reset_index()

salary
```

```python
#Visualización del salario promedio por nivel educativo:

plt.figure(figsize=(10, 6))
sns.barplot(data=salary, x='Salary', y='Education', palette='viridis') 

plt.title('Distribución del salario por nivel educativo')
plt.xlabel('Salario Promedio')
plt.ylabel('Nivel Educativo')
plt.grid(True)
plt.tight_layout()
plt.show()

#Esto nos muestra que a más nivel educativo más alto es el salario promedio, siendo el más bajo el instituto (o escuela obligatoria) y el más alto un doctorado.
```

```python
#5. ¿Cuál es la proporción de clientes con diferentes tipos de tarjetas de fidelidad?

fidelity = clientes['Loyalty Card'].value_counts()

fidelity
```

```python
# Gráfico de pastel para visualizar la proporción:

plt.figure(figsize=(8,6))
plt.pie(fidelity.values, labels=fidelity.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Pastel2.colors, wedgeprops={'edgecolor': 'white', 'linewidth': 2})
plt.title('Distribución de Clientes por Carta de fidelidad')
plt.axis('equal') 
plt.show()

#Esto nos muestra que el tipo de fidelidad más popular es el STAR con un 45,6% de l@s clientes. 
```
